```python
x=df.iloc[:,2:30]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(x)
forest_norm = sc.transform(x)
forest_norm
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 26)
pca_values = pca.fit_transform(forest_norm)
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.figure(figsize=(12,4))
plt.plot(var1,color="green",marker = "P");
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:24],columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11','pc12','pc13','pc14','pc15','pc16','pc17','pc18','pc19','pc20','pc21','pc22','pc23','pc24']),df[['size_category']]], axis = 1)
finalDf.size_category.replace(('large','small'),(1,0),inplace=True)
finalDf
```

```python
array = finalDf.values
x=array[:,0:24]
y=array[:,24]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
array
```

```python
#initialize model

ann=Sequential()
```

```python
#add hidden layer
ann.add(Dense(units=12,activation='relu', input_dim=24))
ann.add(Dense(units=8,activation='relu'))

#Add output layer
ann.add(Dense(units=1,activation='sigmoid'))

#establish connection
ann.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python

#train the model
ann.fit(xtrain,ytrain,epochs=50, validation_split=0.3, batch_size=10,)
```

```python
ypred=ann.predict(xtest)
ypred=ypred>0.5
```

```python
print(classification_report(ytest,ypred))
```
