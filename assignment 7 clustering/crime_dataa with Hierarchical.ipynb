```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('crime_dataa.csv')
df.head()
```

# EDA

```python
x=df.drop(columns=['Unnamed: 0'])
x.head()
```

```python
x.info
```

```python
x.describe()
```

```python
x.isna().sum()
```

```python
from sklearn.preprocessing import normalize
```

```python
norm=pd.DataFrame(normalize(x),columns=x.columns)
norm
```

#  Model Building With Hierarchical

```python
from scipy.cluster import hierarchy
```

```python
lk=hierarchy.linkage(norm,method='complete')
plt.figure(figsize=(10, 7))  
dendrogram=hierarchy.dendrogram(lk)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=5)
y=hc.fit_predict(norm)
```

```python
x['cluster']=y
x
```

```python
x.groupby('cluster').agg('mean')
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(x['cluster'],x['Rape'], c=hc.labels_) 
```
