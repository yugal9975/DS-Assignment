# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x1=sc.fit_transform(x)
x1
```

# Model Building With DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN()
```

```python
db.fit(x1)
```

```python
#checking outliers label -1 
db.labels_
```

# Hyper Parameter Tuning

```python
db=DBSCAN(eps=1,min_samples=12)
db.fit_predict(x1)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x1,db.labels_)
```

```python
# Adding Clusters in dataset
x['Clusters']=db.labels_
x
```

```python
#Plot Clusters for look how many clusters
plt.figure(figsize=(10, 7))  
plt.scatter(x['Clusters'],x['Balance'], c=db.labels_)
```
