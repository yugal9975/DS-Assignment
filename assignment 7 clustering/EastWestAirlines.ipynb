#  Feature Scalling

```python

from sklearn.preprocessing import normalize
```

```python
norm=pd.DataFrame(normalize(df2),columns=df2.columns)
norm
```

```python
from scipy.cluster import hierarchy
```

```python
lk=hierarchy.linkage(norm,method='complete')
dendrogram=hierarchy.dendrogram(lk)
```

```python
lk=hierarchy.linkage(norm,method='average')
dendrogram=hierarchy.dendrogram(lk)
```

# Model Building

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,random_state=2)
    km.fit(norm)
    wcss.append(km.inertia_)
```

```python
#plot elbow chart
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel(' K values')
plt.ylabel('wcss')
plt.show()
```

```python
#create a model with k value is 4

km4=KMeans(4,random_state=30).fit(df)
km4
```

```python
km4.labels_
```

```python
#assign clusters to data set
df3=df2.copy()
df3['y']=km4.labels_
df3
```

```python
km4.cluster_centers_
```

```python
df3.groupby('y').agg(['mean']).reset_index()
```

```python
#Plot Clusters for look how many clusters
plt.scatter(df3['y'],df3['Balance'], c=km4.labels_) 
```
