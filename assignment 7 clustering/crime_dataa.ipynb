# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(df)
x
```

# Model Building

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN()
```

```python
db.fit(x)
```

```python
#checking outliers label -1 
db.labels_
```

# Hyper Parameter Tuning

```python
db=DBSCAN(eps=1,min_samples=4)
db.fit_predict(x)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,db.labels_)
```

```python
# Adding Clusters in dataset
df['Clusters']=db.labels_
df
```

```python
#Plot Clusters for look how many clusters
plt.figure(figsize=(10, 7))  
plt.scatter(df['Clusters'],df['UrbanPop'], c=db.labels_)
```
