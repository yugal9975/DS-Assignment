```python
#  model building

model=smf.ols('Price~Age+KM+HP+CC+Doors+QT+Weight',data=df).fit()
model
```

```python
model.summary()
```

# calculating VIF

```python
v_age=smf.ols('Age~KM+HP+CC+Doors+QT+Weight',data=df).fit().rsquared
vif_age=1/(1-v_age)

v_km=smf.ols('KM~Age+HP+CC+Doors+QT+Weight',data=df).fit().rsquared
vif_km=1/(1-v_age)

v_hp=smf.ols('HP~Age+KM+CC+Doors+QT+Weight',data=df).fit().rsquared
vif_hp=1/(1-v_age)

v_cc=smf.ols('CC~Age+KM+HP+Doors+QT+Weight',data=df).fit().rsquared
vif_cc=1/(1-v_age)

v_Doors=smf.ols('Doors~Age+KM+HP+CC+QT+Weight',data=df).fit().rsquared
vif_Doors=1/(1-v_age)

v_QT=smf.ols('QT~Age+KM+HP+CC+Doors+Weight',data=df).fit().rsquared
vif_QT=1/(1-v_age)

v_weight=smf.ols('Weight~QT+Age+KM+HP+CC+Doors',data=df).fit().rsquared
vif_weight=1/(1-v_age)
```

```python
vif=pd.DataFrame({
    'Features' : ['Age','KM','HP','CC','Doors','QT','Weight'],
    'vif'      : [vif_age,vif_km,vif_hp,vif_cc,vif_Doors,vif_QT,vif_weight]
})
vif
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') 
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>6000))
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(df)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python

(np.argmax(c),np.max(c))
```
