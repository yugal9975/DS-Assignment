```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf

```

```python
dfa=pd.read_csv('50_Startups.csv')
dfa
```

```python
dfa.drop(columns=['State'],inplace=True)
```

```python
dfa
```

```python
df=dfa.rename({'R&D Spend':'RD','Administration':'ADM','Marketing Spend':'MKR'},axis=1)
df
```

```python
sns.heatmap(df.corr(),annot=True,cmap='rainbow')
```

```python
#calculate vif

r_rd=smf.ols('RD~ADM+MKR',data=df).fit().rsquared
v_rd=1/(1-r_rd)

r_adm=smf.ols('ADM~RD+MKR',data=df).fit().rsquared
v_adm=1/(1-r_adm)

r_mkr=smf.ols('MKR~ADM+RD',data=df).fit().rsquared
v_mkr=1/(1-r_mkr)
```

```python
vif=pd.DataFrame({
    'data':['RD','ADM','MKR'],
    'vif' :[v_rd,v_adm,v_mkr]
})
vif
```

```python
#no multicolinarity problem
```

```python
#model bulding

model=smf.ols("Profit~RD+ADM+MKR",data=df).fit()
model
```

```python
model.summary()
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(df)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```
